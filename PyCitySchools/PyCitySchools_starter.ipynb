#Observations:
1. Charter schools have an overwhelmingly higher pass rate. The low performance of District schools though can be strongly attributed to math scores.
>-The size of the school appears to be inversly related to pass rate 
2. The amount of spending per student does not appear to have a positive relationship to the passing rate of the school.
3. A student's reading and math scores are consistent going from grade to grade.

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load, encoding="utf-8")
student_data = pd.read_csv(student_data_to_load, encoding="utf-8")

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#total number of schools
total_schools = len(school_data_complete["school_name"].unique())

#total number of students
total_students = len(school_data_complete["student_name"])

#total budget
total_budget = school_data["budget"].sum()

#average math score
avg_math_score = school_data_complete["math_score"].mean()

#average reading score
avg_reading_score = school_data_complete["reading_score"].mean()

#percentage of students with a passing math score
student_math_passing_total = len(school_data_complete.loc[school_data_complete["math_score"] >= 70])

student_math_passing_percent = (student_math_passing_total/total_students)*100

#percentage of students with a passing reading score
student_read_passing_total = len(school_data_complete.loc[school_data_complete["reading_score"] >= 70])
student_read_passing_percent = (student_read_passing_total/total_students)*100

#percentage of students who passed math **and** reading
student_total_passing = len(school_data_complete.loc[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)])
student_total_passing_percent = (student_total_passing/total_students)*100

#dataframe to hold the above results
df_student_metrics = pd.DataFrame({
    "Total Number of Schools": [total_schools],
    "Total Number of Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [avg_math_score],
    "Average Reading Score": [avg_reading_score],
    "Math Passing Percentage": [student_math_passing_percent],
    "Reading Passing Percentage": [student_read_passing_percent],
    "Overall Passing Percentage": [student_total_passing_percent]
})

display(df_student_metrics)
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#School Name
school_names = school_data_complete["school_name"].unique()

#School Type
school_type = school_data.set_index(["school_name"])["type"]
```

```python
#Total Students
school_total_students = school_data_complete["school_name"].value_counts()
```

```python
#Total School Budget

school_total_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
```

```python
#School Per Student Budget

school_per_student_budget = school_total_budget/school_total_students
```

```python
#average math score by school

school_avg_math_score = school_data_complete.groupby(["school_name"]).mean()["math_score"]
```

```python
#Calculate the average reading score by school
school_avg_reading_score = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
```

```python
#percentage of students with a passing math score by school
school_math_passing_df = school_data_complete.loc[school_data_complete["math_score"] >= 70]
school_math_passing_total = school_math_passing_df.groupby(["school_name"]).count()["student_name"]

school_math_passing_percent = (school_math_passing_total/school_total_students)*100
```

```python
#percentage of students with a passing reading score by school

school_reading_passing_df = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
school_reading_passing_total = school_reading_passing_df.groupby(["school_name"]).count()["student_name"]

school_reading_passing_percent = (school_reading_passing_total/school_total_students)*100
```

```python
#% Overall Passing by school
school_total_passing_df = school_data_complete.loc[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
school_total_passing = school_total_passing_df.groupby(["school_name"]).count()["student_name"]

school_total_passing_percent = (school_total_passing/school_total_students)*100
```

```python
#A dataframe to hold the above results
df_school_metrics = pd.DataFrame({
    "School Names": school_names,
    "School Type" : school_type,
    "Total Students": school_total_students,
    "Budget Per Student": school_per_student_budget,
    "Total Budget" : school_total_budget,
    "Average Math Score": school_avg_math_score,
    "Math Pass Rate": school_math_passing_percent,
    "Average Reading Score": school_avg_reading_score,
    "Reading Pass Rate": school_reading_passing_percent,
    "Overall Pass Rate": school_total_passing_percent    
})

df_school_metrics
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#creating spending range bins
spending_ranges_bins =[0, 584.9, 629.9, 644.9, 679.9]

#subsequent range labels
per_student_spending_ranges = ["<$585","$585-630","$630-645","$645-680"]

df_school_metrics["Spending Ranges(Per Student)"] = pd.cut(df_school_metrics["Budget Per Student"], spending_ranges_bins, labels=per_student_spending_ranges, include_lowest=False)

per_student_spending_grouped = df_school_metrics.groupby("Spending Ranges(Per Student)")

per_student_spending_grouped[["Average Math Score", "Average Reading Score", "Math Pass Rate", "Reading Pass Rate", "Overall Pass Rate"]].mean().round(decimals=2)

```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
#create school size bins
school_size_bins = [0, 999, 1999, 4999]
#subsequent range labels
school_size_ranges = ["Small(<1000)", "Medium(1000-2000)","Large(2000-5000)"]

df_school_metrics["School Size"] = pd.cut(df_school_metrics["Total Students"], school_size_bins, labels=school_size_ranges, include_lowest= False)

school_size_grouped = df_school_metrics.groupby("School Size")

school_size_grouped[["Average Math Score", "Average Reading Score", "Math Pass Rate", "Reading Pass Rate", "Overall Pass Rate"]].mean()
```
